## Data Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("data.csv")
```

```python
data
```

```python
data.drop(['id','Unnamed: 32'],axis=1)
```

```python
import plotly.graph_objs as go
import plotly.offline as py
null_feat = pd.DataFrame(len(data['radius_mean']) - data.isnull().sum(), columns = ['Count'])
plt.figure(figsize=(8,4))
plt.yticks(fontsize=12)
plt.xticks(rotation='vertical',fontsize=8)
xc=plt.bar(x = null_feat.index, height = null_feat['Count'],color='#d3d3d3',edgecolor='k')
                    
plt.savefig("missing_values.png")
```

```python
data['diagnosis']=data["diagnosis"].map({'M':1,'B':0})
```

```python
M = data[(data['diagnosis'] == 1)]
B = data[(data['diagnosis'] == 0)]
M
```

```python
plt.figure(figsize=(7,3))
ax=sns.barplot(x=[len(M),len(B)],y=['M','B'], palette="rocket", saturation=.5)
plt.show()
plt.savefig("MB.png")
```

```python
import matplotlib.pyplot as plt
labels = 'M','B'
sizes = [len(M),len(B)]
explode = (0.1, 0)
plt.pie(sizes, explode=explode, labels=labels,shadow=True, colors=['salmon','#ADD8E6'],startangle=90)
plt.show()
```

```python
def distribution(feature):
    c=sns.distplot(M[feature], color="y",label="Malignant")
    cn=sns.distplot(B[feature],color='b',label="Benign")
    plt.legend()
    plt.show()
```

```python
distribution("radius_mean")
distribution('texture_mean')
distribution('perimeter_mean')
distribution('area_mean')
distribution('smoothness_mean')
distribution('compactness_mean')
distribution('concavity_mean')
distribution('concave points_mean')
distribution('symmetry_mean')
distribution('fractal_dimension_mean')
```

```python
def plot_feat1_feat2(feat1, feat2) :  
    trace0 = sns.scatterplot(x = M[feat1],y = M[feat2],label="Malignant")
    trace1 = sns.scatterplot(x = B[feat1],y = B[feat2],label="Benign")
    plt.xlabel(feat1)
    plt.legend()
    plt.ylabel(feat2)
    plt.show()
    
```

## Positive Correlated Features

```python
plot_feat1_feat2('perimeter_mean','radius_worst')
plot_feat1_feat2('area_mean','radius_worst')
plot_feat1_feat2('texture_mean','texture_worst')
plot_feat1_feat2('area_worst','radius_worst')
```

## Uncorrelated Features

```python
plot_feat1_feat2('smoothness_mean','texture_mean')
plot_feat1_feat2('radius_mean','fractal_dimension_worst')
plot_feat1_feat2('texture_mean','symmetry_mean')
plot_feat1_feat2('texture_mean','symmetry_se')
```

## Negative Correlation

```python
plot_feat1_feat2('area_mean','fractal_dimension_mean')
plot_feat1_feat2('radius_mean','fractal_dimension_mean')
plot_feat1_feat2('area_mean','smoothness_se')
plot_feat1_feat2('smoothness_se','perimeter_mean')
```
